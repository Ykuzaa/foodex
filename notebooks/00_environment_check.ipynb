# Environment Check

Execute cette cellule avant le notebook d'entrainement. Si elle echoue, utilise le script `scripts/setup_foodex_env.sh` depuis un terminal Jupyter.

```python
from pathlib import Path
import sys

def find_project_root(start=None):
    start = Path.cwd() if start is None else Path(start)
    for candidate in [start, *start.parents]:
        if (candidate / 'scripts' / 'verify_environment.py').exists():
            return candidate
    raise FileNotFoundError('Lance ce notebook depuis foodex/ ou foodex/notebooks/.')

project_root = find_project_root()
print('Project:', project_root)
print('Python:', sys.version)

%run "{project_root / 'scripts' / 'verify_environment.py'}"
```
